```python
import matplotlib.pyplot as plt
a = [230,560,780,127,128]
b = [200,160,270,127,400]
years = [1,2,3,4]
profit_a = [(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b = [(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth = '3',label =
'CompanyA',marker='>',ms='15',mec='k')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label =
'CompanyB',marker = 'H')
plt.show()
```

subplot function is used to seperate the companies or graph
in one graph ther is 2 companies or graph

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
print(iris)
sns.jointplot(x="sepal_length",y="petal_width",data=iris,label="amulya")
plt.title("Scatter plot of total bill vs tip")
plt.xlabel("sepal($)")
plt.xlabel("petal($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
print(iris)
sns.countplot(x="sepal_length",data=iris,label="amulya")
plt.title("Scatter plot of total bill vs tip")
plt.xlabel("sepal($)")
plt.xlabel("petal($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
print(tips)
sns.lmplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter plot of total bill vs tip")
plt.xlabel("totalbill($)")
plt.ylabel("Tips($)")
plt.show()
```
